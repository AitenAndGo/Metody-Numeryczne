Interpolacja jest to metoda określania funkcji, która na podstawie dostarczonych punktów szacuje wartości pośredniczące między nimi. Najprostrzym przykładem interpolacji jest interpolacja liniowa, gdzie pomiędzy dwoma kolejnymi punktami tworzy się linię. Są też inne bardziej zaawasnowane metody interpolacji jak interpolacja kwadratowa, gdzie koljene wartości przybliża się korzystając z funkcji kwadratowej. Są też interpolacje wielomianowe o wyższym stopniu wielomianu. Istnieją również interpolacje typu spline, które są fukcjami sklejanymi, są to wygładzone funkcje wielomianowe często stosowane w danych, gdzie dochodzi do łagodnych przejść pomiędzy punktami. Interpolacja jest potrzebna w sytuacji gdzie po dokonaniu pomiarów chcemy oszacować jakie wartości wyspępują w miejscach, w których nie wykonaliśmy próbkowania.

## 2. Zadania
<div style="text-align: justify; line-height: 1.6; margin: 10px;">
W następnej części sprawozdania zostaną przedstawione różne rodzaje interpolacji. Całość wykonana w środowisku jupiter notebook przy pomocy biblioteki pandas.
Na początku importujemy biblioteki oraz tworzymy dane syntetyczne w tym przypadku postanowiłem na wygenerowanie danych na podstawie funkcji y = x z tym że do każdego z punktów dodałem losowy szum z przedziału -10 do 10.
</div>

```python

import numpy as np
import random
import pandas as pd
```

```python

# listy z punktami z przedziału od 0 do 10 z krokiem 0.1
Points_x = np.arange(0, 10, 0.1)
Points_y = [None] * len(Points_x)

# ustawiamy co 10 wartość z listy na f(x) = x + (odchyłka)
for i in range(0, len(Points_x), 10):
    # odchyłka z przedziału -10 : 10 aby było dobrze widoczne na wykresie
    error = random.uniform(-10, 10)
    Points_y[i] = Points_x[i] + error
```

<div style="text-align: justify; line-height: 1.6; margin: 10px;">
Następnie tworzymy tabele danych z naszymi punktami i tworzymy różne interpolacje uzupełniając dane z krokiem co 0,1. Jak widać na poniższym wykresie przedstawiono interpolacje liniową, kwadratową, sześcienną, wielomianową i schodkową. 
</div>


```python
# tworzymy nową tabele danych z naszymi wartościami x i y
df = pd.DataFrame(data={"value": Points_y}, index=Points_x)

# tworzymy różne interpolacje na naszych danych
linear = df.interpolate(method='linear')
quadratic = df.interpolate(method='quadratic')
cubic = df.interpolate(method='cubic')
polynomial = df.interpolate(method='polynomial', order=5)
nearest = df.interpolate(method='nearest')

# towrzymy słownik aby przypisać każdej z interpolacji jej nazwę
interpolations = {
    "liniowa": linear,
    "kwadratowa": quadratic,
    "sześcienna": cubic,
    "wielomianowa": polynomial,
    "schodkowa": nearest
}

# tworzymy nowe colum ny danych z odpowiednimi interpolacjami
for name, interpolation in interpolations.items():
    df[name] = interpolation['value']

# tworzymy wykres ze wszytskimi typami interpolacji
ax = df.loc[0:9].plot(y=['liniowa', 'kwadratowa', 'sześcienna', 'wielomianowa', 'schodkowa'], 
             figsize=(10, 6), grid=True, title="Interpolacje")

ax.set_xlabel("X")
ax.set_ylabel("Y")
```

## 3. Wnioski
<div style="text-align: justify; line-height: 1.6; margin: 10px;">
Analizując powyższy wykres można zauważyć różnice jakie wyspępują przy różnych rodzajach interpolacji. W przedstawionym przypadku najlepiej dopasowaną wydaje się być interpolacja kwadratowa, jednak jest to ściśle związane z danymi jakie posiadamy. Interpolacja wielomianowa bardzo dobrze odzwierciedla rzeczywistość w środkowej części wykresu, natomiast na początku i końcu posiada dość duże wybiegi, sześcienna interpolacja różnież posiada większe wybiegi na początku i końcu, natomiast kwadratowa wydaje się być najlepiej dopasowana w tym przypadku. Nie oznacza to że w każdym użytym zestawie danych i różnych opisach pomiarów najleszy jest tylko jeden rodzaj interpolacji. Każda z przedstawionych może być przydatna przy różnych zastosowaniach.
