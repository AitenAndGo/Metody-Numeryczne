## 2. Zadania

### A. Wczytanie danych przykładowych

Plik z danymi został pobrany z przygotowanych na stronie [danych przykładowych](https://temisto.immt.pwr.wroc.pl/~myszka/Dane_Numeryczne/netatmo/?C=N;O=A), w tym przypadku wybrany został plik z informacjami o zmianie temperatury, który jest dołączony w folderze z plikiem głównym. Pobrany plik jest typu .csv (comma seperated value). Otwierając ten plik w edytorze można zauważyć, że na początku znajdują się dane dotyczące: nazwy stacji, pozycji geograficznej, nazwy modułu oraz typu modułu. Poniżej znajdują się już dane w trzech kolumnach. W pierwszej jest timestamp, jest to liczba sekund jaka upłyneła od rozpoczęcia pomiaru czasu w ten sposób na komputerach czyli od początku 1970 roku. W drugiej kolumnie jest dokładna data wykoniania pomiaru oraz w ostatniej kolumnie wynik samego pomiary temperatury w stopniach Celsjusza.


W celu obsłużenia pliku będzie nam potrzebna biblioteka csv. Następnie otwieramy plik, zaznaczając że w tym pliku kolumny są rozdzelane średnikiem a nie przecinkiem jak jest to domyślnie. Przechodząc po wszytkich wierszach możemy je wypisać aby zobaczyć co się tam kryje.

```python
import csv

# Otwarcie pliku csv
with open('Outdoor_temperature_2017-09.csv', 'r') as csvfile:
  
  # Stworzenie obiektu odczytującego, ustawienie seperatora jako średnika
  csv_reader = csv.reader(csvfile, delimiter=';')
  
  # Iteracja po 10 pierwszych wierszach wraz z ich wypisaniem
  for i, wiersz in enumerate(csv_reader):
    print(wiersz)
    if i == 10:
        break
```

Aby móc później łatwo się posługiwać przedstawoinymi danymi zapiszemy interesujace nas dane takie jak timestamp, datę, oraz pomiar w osobnych listach. Wykorzystamy bibliotekę pandas która pozwala na tworzenie tablic danych co ułatwi prace i tworzenie wykresów.

```python
import pandas as pd
import numpy as np

# wczytanie pliku z uwzględnieniem średnika jako seperatora, oraz pominięcie pierwszych 4 wierszy, które nie posiadają intresujących nas danych
df = pd.read_csv('Outdoor_temperature_2017-09.csv', delimiter=';', skiprows=4)

# zapisa odpowienich danych z kolumn w listy, 
timeStamp = df.iloc[:, 0].tolist()
data = df.iloc[:, 1].tolist()
temperatura = df.iloc[:, 2].tolist()

# Wyświetlenie pierwszych kilku wierszy
df.head()
```

### B. Narysowanie wykresu danych

W celu stworzenia wykresu posłużę się biblioteką matplotlib. W dodatku usuniemy z naszych danych wszystkie elemety które nie posiadają danych (w aktualnym pliku nie ma takich danych, natomiast w innych plikach istnieją dane które posiadają NaN (Not a Number), które należy usunąć)

```python
# iteracja po wszystkich wczytanych danych
for i in range(len(temperatura)):
    # jeżeli temperatura nie została odczytana usuwamy te dane ze wszytkich kolumn
    if temperatura[i] == "NaN":
        temperatura.pop(i)
        timeStamp.pop(i)
        data.pop(i)
```

```python
# wybieramy interesujący nas przedział dat
# Ustawienia zakresu dat (a i b) jako stringi
a = '2017/09/02 00:00:00'
b = '2017/09/03 23:59:59'
```

```python
df = pd.DataFrame(data={"data": data, "temperatura": temperatura})

# Konwersja kolumny z datami do specjalnego typu datetime
df['data'] = pd.to_datetime(df['data'])

# Ustawienie kolumny 'data' jako indeks
df.set_index('data', inplace=True)

# Konwersja kolumny 'temperatura' na typ float
df['temperatura'] = df['temperatura'].astype(float)

# wyświetlenie pierwszych kilku wierszy
df.head()
```

```python
# Rysowanie wykresu z wcześniej ustalonego przedziału dat
df[a:b].plot()
```

### C. Zaznaczenie na wykresie okresów wzrostu i spadku mierzonej wartości

Aby zaznaczyć wzrosty i spadki temeratury posłużę się zróżniczkowanym wykresem, będzie wtedy dobrze widoczny przyrost temeratury na przedziale czasowym

```python
# obliczenie pohodnej temperatury
df['przyrost temperatury'] = df['temperatura'].diff()

# narysowanie wykresu temperatury(lewo, niebieski), oraz przyrostu temperatury(prawo, pomarańczowy)
df[a:b].plot(y=['temperatura', 'przyrost temperatury'], secondary_y='przyrost temperatury')
```

Teraz aby wizualnie to lepiej wyglądało ustawię wykres w miejscach rosnącej, stałej i majelącej temperaturze na inne kolory.

```python
wzrost = []
spadek = []
stala = []

# iteracja po wszystkich temperaturach
for i, temp in enumerate(df['przyrost temperatury']):
    # jeżeli pohodana większa od zera to temperatura rośnie
    if temp > 0:
        wzrost.append(df['temperatura'].iloc[i])
        spadek.append(np.nan)
        stala.append(np.nan)
    # jeżeli pohodana większa od zera to temperatura maleje
    elif temp < 0:
        wzrost.append(np.nan)
        spadek.append(df['temperatura'].iloc[i])
        stala.append(np.nan)
    # w innym przypadku temeratura nie zmienia wartości
    else:
        wzrost.append(np.nan)
        spadek.append(np.nan)
        stala.append(df['temperatura'].iloc[i])

# Dodanie kolumn do ramki danych
df['wzrost temperatury'] = wzrost
df['spadek temperatury'] = spadek
df['stala temperatura'] = stala

# narysownie wykresu
df[a:b].plot(y=['wzrost temperatury', 'spadek temperatury', 'stala temperatura'])
```

### D. Wygenerowanie „syntetycznych" zestawów danych i sprawdzenie procedur całkowania/różniczkowania.

Na początku zajmę się generacją syntetycznych danych, są to dane losowe.

```python
# Ustawienie ziarna
np.random.seed(0)

# generacja 100 losowych liczb (danych) w zakresie od -1 do 1
data = np.random.uniform(low=-1, high=1, size=100)
df = pd.DataFrame(data, columns=['dane'])

# wykres
df.plot()
```

Teraz kolejne wykresy z całkowaniem i różniczkowaniem wykresu

```python
df['różniczka'] = df['dane'].diff()
df['całka'] = df['dane'].cumsum()

df.plot()
```

### E. Wygenerowanie syntetycznych zestawów danych (zaburzonych szumem o rozkładzie normalnym o średniej zero i  zadanej wariancji) oraz sprawdzenie procedur całkowania/różniczkowania i zbadanie wpływu wariancji na wyniki.

Do wcześniejszych danych dodajemy szum o rozkładzie normalnym i zadanej wariancji 


```python
# ustawienie wariancji na wartość 2.5
w = 2.5

# generacja szumu o rozkładzie normalnym, wartości śreniej równej 0(loc=0), odchyleni
szum = np.random.normal(loc=0, scale=w**0.5, size=100)

# powiększenie danych o szum
df['dane'] = df['dane'] + szum

# obliczenie różniczki oraz całki
df['różniczka'] = df['dane'].diff()
df['całka'] = df['dane'].cumsum()

# wykres
df.plot()
```

A teraz zobaczymy jaki wpływ będzie miała zmiana wariancji

```python
# Ustawienie ziarna
np.random.seed(0)

# dane syntetyczne (sinus)
x = np.linspace(0, 5 * 2 * np.pi, 100)
data = np.sin(x)
df = pd.DataFrame(data, columns=['dane'])

for i in range(0, 5, 1):
    w = i/10
    # generacja szumu o rozkładzie normalnym, wartości śreniej równej 0(loc=0), odchyleni
    szum = np.random.normal(loc=0, scale=(w)**0.5, size=100)

    # powiększenie danych o szum
    df['dane dla w = ' + str(w)] = df['dane'] + szum

    # obliczenie różniczki oraz całki
    df['różniczka dla w = ' + str(w)] = df['dane dla w = ' + str(w)].diff()
    df['całka dla w = ' + str(w)] = df['dane dla w = ' + str(w)].cumsum()

    # wykres
    df.plot(y=['dane dla w = ' + str(w), 'różniczka dla w = ' + str(w), 'całka dla w = ' + str(w)])
```
